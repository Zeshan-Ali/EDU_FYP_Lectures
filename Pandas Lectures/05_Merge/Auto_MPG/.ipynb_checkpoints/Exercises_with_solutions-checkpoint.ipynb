```python
import pandas as pd
stu=pd.read_csv("stu.csv")
```

```python
stu.head()
```

```python
def percentage(x):
    if x > 70:
        return "A"
    else:
        return "B"
stu['perc'] = stu['Phy'].apply(percentage)
stu.head()
```

```python
def perc(r):
    p = (((r['Bio'] + r['Chem'] + r['Phy'])*100) / 300)
    return p
```

```python
stu['perc'] = stu.apply(perc, axis=1)
```

```python
stu.head()
```

```python
def grade(p):
    if p > 80:
        return "A"
    elif p >70:
        return "B"
    elif p >60:
        return "C"
    elif p >50:
        return "D"
    elif p >40:
        return "E"
    else:
        return "F"
    
```

```python
stu['g'] = stu['perc'].apply(grade)
```
